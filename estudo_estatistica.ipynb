```python
# variável independente (que explica a variável dependente)

X = df['Total Bags'].values

X
```

```python
# variável dependente (é explicada pela variável independente)

y = df['Total Volume'].values

y
```

```python
# gráfico de dispersão entre x e y

plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = X.reshape(-1, 1)
y = y

modelo = LinearRegression()

# treinar o modelo pra aprender a prever a variável y
modelo.fit(X, y)
```

```python
plt.scatter(X, y)

# reta de regressão (pegar o X e y previsto)
plt.plot(X, modelo.predict(X), color='hotpink')
```

```python
bags = 76
modelo.predict([[bags]])
```

```python
# verificar o quão confiável é o modelo
rsquared = modelo.score(X, y)
rsquared
```

```python
# verificar o quão confiável as previsões são
from sklearn import metrics
```

```python
metrics.r2_score(y_true = y, y_pred = modelo.predict(X))
```
