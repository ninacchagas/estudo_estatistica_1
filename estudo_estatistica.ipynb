```python
# variÃ¡vel independente (que explica a variÃ¡vel dependente)

X = df['Total Bags'].values

X
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = X.reshape(-1, 1)
y = y

modelo = LinearRegression()

# treinar o modelo pra aprender a prever a variÃ¡vel y
modelo.fit(X, y)
```

```python
plt.scatter(X, y)

# reta de regressÃ£o (pegar o X e y previsto)
plt.plot(X, modelo.predict(X), color='hotpink')
```

```python
bags = 76
modelo.predict([[bags]])
```

```python
# verificar o quÃ£o confiÃ¡vel Ã© o modelo
rsquared = modelo.score(X, y)
rsquared
```

```python
# verificar o quÃ£o confiÃ¡vel as previsÃµes sÃ£o
from sklearn import metrics
```

```python
metrics.r2_score(y_true = y, y_pred = modelo.predict(X))
```
